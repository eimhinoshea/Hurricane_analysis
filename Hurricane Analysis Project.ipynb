```python
#A function that returns a new list of updated damages where the recorded data is converted to float values and the missing data is retained as "Damages not recorded".
damages_float = []
def string_to_float(lst):
  for record in damages:
    if record[-1] == "M":
      damages_float.append((float(record[:-1]) * 1000000))
    elif record[-1] == "B":
      damages_float.append((float(record[:-1]) * 1000000000))
    else:
      damages_float.append(record)
  return damages_float
  
print(string_to_float(damages))
```

```python
#a function that constructs a dictionary made out of the lists, where the keys of the dictionary are the names of the hurricanes, and the values are dictionaries themselves containing a key for each piece of data about the hurricane.
hurricane = {}
def hurricane_dict(names,months,years,max_sustained_winds,areas_affected,damages,deaths):
  for i in range(len(names)):
    hurricane[names[i]] = {
      "Name": names[i],
      "Month": months[i],
      "Year": years[i],
      "Max Sustained Wind": max_sustained_winds[i],
      "Areas Affected": areas_affected[i],
      "Damage": damages_float[i],
      "Deaths": deaths[i]}
  return hurricane

hurricane_dict(names,months,years,max_sustained_winds,areas_affected,damages,deaths)
```

```python
# create a new dictionary of hurricanes with year as key
def hurricane_year(dict1):
  hurricane_by_year = {}
  for i in dict1:
    hurricane_by_year[dict1[i]["Year"]] = dict1[i]
  return hurricane_by_year

hurricane_year(hurricane)
```

```python
#Counting Damaged Areas
count_of_affected_areas = {}
def area_affected_count(dict1):
  for i in dict1:
    for i in dict1[i]["Areas Affected"]:
      if i not in count_of_affected_areas:
        count_of_affected_areas[i] = 1
      else:
        count_of_affected_areas[i] = count_of_affected_areas[i] + 1
  return count_of_affected_areas
    
area_affected_count(hurricane)
```

```python
# find highest mortality hurricane and the number of deaths

def most_deaths(dict1):
  highest_death_hurricane = list(dict1.keys())[0]
  highest_deaths = dict1[highest_death_hurricane]["Deaths"]
  for i in dict1:
    if dict1[i]["Deaths"] > highest_deaths:
      highest_deaths = dict1[i]["Deaths"]
      highest_death_hurricane = dict1[i]["Name"]
  return "The hurricane which caused the greatest number of deaths is " + highest_death_hurricane + " with " + str(highest_deaths) + " deaths."

print(most_deaths(hurricane))
```

```python
# Rating Hurricanes by Mortality
def mortality_rating(dict1):
  hurricanes_by_mortality = {0:[],1:[],2:[],3:[],4:[],5:[]}
  for i in dict1:
    if dict1[i]["Deaths"] == 0:
      hurricanes_by_mortality[0].append(dict1[i])
    elif 0 < dict1[i]["Deaths"] <= 100:
      hurricanes_by_mortality[1].append(dict1[i])
    elif 100 < dict1[i]["Deaths"] <= 500:
      hurricanes_by_mortality[2].append(dict1[i])
    elif 500 < dict1[i]["Deaths"] <= 1000:
      hurricanes_by_mortality[3].append(dict1[i])
    elif 1000 < dict1[i]["Deaths"] <= 10000:
      hurricanes_by_mortality[4].append(dict1[i])
    elif 10000 < dict1[i]["Deaths"]:
      hurricanes_by_mortality[5].append(dict1[i])
    else:
      print("Number not in scale")
  return hurricanes_by_mortality

print(mortality_rating(hurricane))
```
